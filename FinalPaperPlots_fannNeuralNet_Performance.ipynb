```python
from donut import *
from math import *
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from pylab import *
%matplotlib inline

import pandas as pd
import seaborn as sns
#cp = ["#d01c8b","#f1b6da","#b8e186","#4dac26"]
cp = ["#8e0152","#f1b6da","#4d9221","#276419"]

sns.palplot(sns.color_palette(cp, 4))
sns.set(style='ticks',palette=cp)
colorm="#8e0152"
colorp="#f1b6da"
colorlg="#4d9221"
colorg="#276419"
```

```python
f = open('/Users/Rylan/cisc/creamTeam/FANN/testData/fann_HLIv4_elliotDeepArch_ElliotOut_randWeights_wkIt_ivL25_5randInterps_2MtestingOutput_1_v2.txt','r')
testData = [(float(line.split()[0]),float(line.split()[1])) for line in f]
print(len(testData))

min = 1
max = 0
match,nonmatch,total = [],[],[]
for i in range(len(testData)):
    if testData[i][0] == 1:
        match.append(1-testData[i][1])
    else:
        nonmatch.append(1-testData[i][1])
    total.append(1-testData[i][1])
    if total[-1] > max:
        max = total[-1]
    if total[-1] < min:
        min = total[-1]
        
print("min = ", min, "\nmax = ", max)
scTotal, scMatch, scNonmatch = [],[],[]
[scTotal.append((total[i]-min)/(max-min)) for i in range(len(total))]
[scMatch.append((match[i]-min)/(max-min)) for i in range(len(match))]
[scNonmatch.append((nonmatch[i]-min)/(max-min)) for i in range(len(nonmatch))]

total = scTotal
match = scMatch
nonmatch = scNonmatch
    
bins = linspace(0,1,100)
eventst, edgest, patchest = hist(total, bins, normed=False, log=True, label="total")
eventsm, edgesm, patchesm = hist(match, bins, normed=False, log=True, label="matches",alpha=0.9)
eventsnm, edgesnm, patchesnm = hist(nonmatch, bins, normed=False, log=True, label="non-matches",alpha=0.8)

probBins = [ (bins[i]+bins[i+1])/2 for i in range(len(bins)-1) ]
probHist = [ float(eventsm[i])/float(eventst[i]) if eventst[i]!=0 else 0 for i in range(len(eventsm))]
eventstL = [ eventst[i] for i in range(len(eventst))]
```

```python
lfs = 19
tfs = 15
legfs = 15
xp = 15
yp = 13

fig, ax1 = plt.subplots(figsize=(10,7))
ax1.set_xlabel("Neural Network Output",fontsize=lfs)
ax1.xaxis.labelpad = xp
ax1.hist(nonmatch,bins,log=True,alpha=1.0,color=colorm,label="Non-Matched Validation Pairs ")
ax1.hist(match,bins,log=True,alpha=0.75,color=colorp,label="Matched Validation Pairs ")
ax1.set_ylabel("Number of Validation Pairs",fontsize=lfs,color=colorm)
ax1.legend(loc="upper right",prop={'size':legfs})
ax1.set_yscale('log')
ax1.yaxis.labelpad = yp
for tl in ax1.get_yticklabels():
    tl.set_color(colorm)
    tl.set_size(tfs)
for tl in ax1.get_xticklabels():
    tl.set_size(tfs)
    
ax2 = ax1.twinx()
ax2.plot(probBins,probHist,linestyle='--',linewidth=3,color=colorlg,label="Ratio",alpha=0.7)
ax2.set_ylabel("Ratio: Matched/All",fontsize=lfs,color=colorg)
ax2.yaxis.labelpad = yp
for tl in ax2.get_yticklabels():
    tl.set_color(colorg)
    tl.set_size(tfs)
    
savefig('/Users/Rylan/cisc/creamTeam/DonutDocs/plots/finalPlots/NN_prob_and_matchedNonMatched_flippeColors.pdf')
plt.show()
```
