```python
from donut import *
import random
import pickle
import math
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.backends.backend_pdf import PdfPages 
%matplotlib inline

import pandas as pd
import seaborn as sns

#cp = ["#8e0152","#de77ae","#f1b6da","#4d9221","#276419"]
cp = ["#c51b7d","#de77ae","#f1b6da","#4d9221","#276419"]
#cp = ["#de77ae","#f1b6da","#b8e186","#7fbc41","#4d9221"]
#cp = ["#c51b7d","#de77ae","#b8e186","#7fbc41","#4d9221"]
sns.palplot(sns.color_palette(cp, 5))

cm = sns.light_palette("#8e0152",as_cmap=True)
sns.set(style='ticks',palette=cp)
```

```python
random.seed()
randSeed = random.randint(0,1000000)

w = cruller.WordList()
w.LoadFromFile("/Users/Rylan/cisc/creamTeam/wordLists/WordList_100000_combined_check.dat")
#w = sprinkles.MostCommon(w,20000)
w.SetSeed(randSeed)

k = bearclaw.MakeStandardKeyboard()
k.SetSeed(randSeed)
k.RemoveKey('.')
kStan = k

ivL = 75
model = []
for i in range(5):
    model.append(cruller.SimpleInterpolationModel())
    model[i].SetXScale(0.75)
    model[i].SetYScale(0.5)
    model[i].SetCorrelation(0.5)
    model[i].SetVectorLength(ivL)
    model[i].SetSeed(random.randint(0,1000000))


model[1].Interpolation = cruller.CubicSplineInterpolation
model[2].Interpolation = cruller.ModCubicSplineInterpolation
model[3].Interpolation = cruller.HermiteCubicSplineInterpolation
model[4].Interpolation = cruller.MonotonicCubicSplineInterpolation
```

```python
word="while"
iv = model[1].RandomVector(word,kStan)
ivl = []
[ivl.append(iv) for i in range(5)]
bearclaw.DrawKeyboard(kStan,inputvector=ivl,frequencymap=word,colormap=mpl.cm.Reds,figsize=(8,4),nopalette=True)
```
