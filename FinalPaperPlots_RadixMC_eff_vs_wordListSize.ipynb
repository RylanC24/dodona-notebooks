```python
from donut import *
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from pylab import *
import random
%matplotlib inline

import pandas as pd
import seaborn as sns
#cp = ["#d01c8b","#f1b6da","#b8e186","#4dac26"]
cp = ["#c51b7d","#f1b6da","#4d9221"]

sns.palplot(sns.color_palette(cp, 3))
sns.set(style='ticks',palette=cp)
colorm="#c51b7d"
colorp="#f1b6da"
colorg="#4d9221"
```

```python
f = open('/Users/Rylan/cisc/creamTeam/donutData/systematics/timingTest.txt','r')
testData = [(float(line.split()[0]),float(line.split()[1])) for line in f]
#[print(testData[i][0],"   ",testData[i][1]) for i in range(len(testData))]
radixTimes = [testData[i][0] for i in range(len(testData))]
mcTimes = [testData[i][1] for i in range(len(testData))]

nWordsL = [95881,70000,50000,40000,30000,20000,15000,10000,5000,3000,1000,500]
errors = [0 for i in range(len(nWordsL))]
```

```python
lfs = 19
tfs = 15
legfs = 15
xp = 6
lw = 3

fig,ax = plt.subplots(figsize=(10,6))
plt.errorbar(nWordsL,radixTimes,yerr=errors,color=colorg,fmt='-o',label='RadixMC',linewidth=lw)
plt.errorbar(nWordsL,mcTimes,yerr=errors,color=colorm,fmt='-o',label='MC',linewidth=lw)
plt.xlim(400,max(nWordsL)+max(nWordsL)/10)
plt.ylim(15, 30000)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('# of words',fontsize=lfs)
plt.ylabel('time (s)',fontsize=lfs)
plt.yticks(fontsize=tfs)
plt.xticks(fontsize=tfs)
ax.xaxis.labelpad = xp
legend(loc=2,prop={'size':legfs})

import os
filename = '/Users/Rylan/cisc/creamTeam/DonutDocs/plots/finalPlots/timeTest_Radix_v_MC.pdf'
if not os.path.exists(filename):
    plt.savefig(filename)
#    plt.close()
else:
    print("stop writing over you're other $tuff")
show()
```
