```python
random.seed()
randSeed = random.randint(0,1000000)

w = cruller.WordList()
w.LoadFromFile("/Users/Rylan/cisc/creamTeam/wordLists/WordList_100000_combined_check.dat")
#w = sprinkles.MostCommon(w,20000)
w.SetSeed(randSeed)

k = bearclaw.MakeStandardKeyboard()
k.SetSeed(randSeed)
k.RemoveKey('.')
kStan = k

ivL = 50
model = []
for i in range(5):
    model.append(cruller.NeuralNetworkModel("/Users/Rylan/cisc/creamTeam/FANN/trainedNets/fann_donutSwypeDistance_HLIv4_elliotDeepArch_SigmoidOut_randWeights_2kIt_ivL50_randInterp.net"))
    model[i].SetXScale(0.75)
    model[i].SetYScale(0.5)
    model[i].SetCorrelation(0.5)
    model[i].SetVectorLength(ivL)
    model[i].SetSeed(random.randint(0,1000000))


model[1].Interpolation = cruller.CubicSplineInterpolation
model[2].Interpolation = cruller.ModCubicSplineInterpolation
model[3].Interpolation = cruller.HermiteCubicSplineInterpolation
model[4].Interpolation = cruller.MonotonicCubicSplineInterpolation
```
