```python
from donut import *
from datetime import datetime
import random
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

#Set up the word list
randSeed = random.randint(0,1000000)
w = cruller.WordList()
w.LoadFromFile("/Users/Rylan/cisc/creamTeam/wordLists/WordList_100000_combined_check.dat")
w = sprinkles.MostCommon(w,20000)
w.SetSeed(randSeed)

#Set up the input model parameters
nn_model_path = '/Users/Rylan/cisc/creamTeam/FANN/trainedNets/fann_donutSwypeDistance_HLIv4_elliotDeepArch_ElliotOut_randWeights_2kIt_ivL50_5randInterps.net'
nMC = 20000
nRadixSteps = 20
ivL = 50
Xscale = 0.5
Yscale = 0.5
corr = 0.0

model = cruller.NeuralNetworkModel(nn_model_path)
model.SetXScale(Xscale)
model.SetYScale(Yscale)
model.SetCorrelation(corr)
model.SetSeed(randSeed)
model.SetVectorLength(ivL)
```

```python
#Helper functions

def draw_keyboard(k, figsize=(12,6)):
    bearclaw.DrawKeyboard(k, wordlist=w, colormap=matplotlib.cm.Reds, figsize=figsize, nopalette=True)
    
def calculate_error_rate(k):
    eff = cruller.RadixMonteCarloEfficiency(k, model, w, nMC, nRadixSteps)
    print('Error Rate = %.3f%% +/- %.3f%%\n' % (100*(1-eff.Fitness()), 100*eff.Error()))
```
