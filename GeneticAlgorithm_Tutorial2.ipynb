# Genetic Algorithm 

The code in the cell immediately below this just 
- imports the relevent modules
- sets up the word list
- sets the input model
- Make a standard qwerty keyboard with the '.' key missing
- Randomizes the standard keyboard and saves Nk random copies as the initial keyboard population
- defines the fitness function
- defines a few helpful functions that are probably unnecessary but I'm not very good with Python yet

```python
from donut import *
import numpy as np
import matplotlib.pyplot as plt
from random import shuffle
from math import sqrt
%matplotlib inline

#import a word list
w = cruller.WordList()
with open("/Users/Rylan/cisc/swypeAwesome/wordLists/HuckFinn.txt","r", errors="replace") as f:
    for line in f:
        for word in line.split():
            word = word.lower()
            word = ''.join(c for c in word if c.isalpha())
            if len(word) != 0:
                w.AddWord(word)

                
#How much of the world list to consider in efficiency calculations
nW2 = 750
w2 = cruller.WordList()
for i in range(w.Words()):
    if i < nW2:
        w2.AddWord(w.Word(i),w.Occurances(i))
                
#Set the input model to a simple gaussian
model = cruller.SimpleGaussianModel()
model.SetScale(2)

#Make a standard keyboard without the '.' key
k = bearclaw.MakeStandardKeyboard()
k.RemoveKey('.')

#Get a list of Nk randomized keyboards (all without the '.' key)
Nk = 10 #number of keyboards in a generation
klist = [ bearclaw.RandomKeyboard(k) for i in range(Nk) ]

#set the number of generations
nGen = 10

#Set the fitness function
Nit=1500 #number of monte carlo trials to run (this is the bottleneck in terms of cpu time)
def MCFitness(kb):
    return cruller.MonteCarloEfficiency(kb,model,w2,Nit)

#return the average fitness value from a generation of keyboards(kList)
def getAve(kl):
    average = 0
    for kb in kl:
        average += kb[1].Fitness()
    return average/len(kl)   

#return the maximum fitness value from a generation of keyboards (kList)
def getMax(kl):
    maxFitness = sorted([kb[1].Fitness() for kb in kl])[-1]
    for kb in kl:
        if kb[1].Fitness() == maxFitness:
            return (kb[0],kb[1])

def getMin(kl):
    minFitness = sorted([kb[1].Fitness() for kb in kl])[0]
    for kb in kl:
        if kb[1].Fitness() == minFitness:
            return (kb[0],kb[1])
```

  
  
**The cell below just sets up a simple plotting function to show the progress of the GA after each generation**
  
  

```python
def draw_progress(x,y,yerr,ymax,ymaxerr, ymin, yminerr):
    plt.figure()
    p1=plt.errorbar(x, y, yerr=yerr, marker='o',color='red')
    p2=plt.errorbar(x, ymax, yerr=ymaxerr, marker='o',color='blue')
    p3=plt.errorbar(x, ymin, yerr=yminerr, marker='o',color='blue')
    plt.xlim((-2,nGen+2))
#    plt.ylim((0.5,1))
    plt.xlabel("Generation")
    plt.ylabel("Fitness")
    l1 = plt.legend([p1,p2], ["Generation Average", "Generation Maximum"], loc=4)
    plt.show()
```

### New Features of GA

The Main difference between the new glazed.Evolve() function and the old one (from the first ipython notebook sent out) is that the first parameter, kList, is no longer a list of (cruller.Keyboard keyboard, double fitnessValue) tuples.  It is now a list of (cruller.Keyboard keyboardObject, cruller.FitnessResult fitnessObject), so the type of the second parameter of the tuple changed.  The reason for this is that the cruller.FitnessResult object stores the associated error along with each fitness value calculation.  

Having the error along with each fitness value is helpful because now it can be used in the Evolve function to give more meaning to the natural selection weighting.  It also allowed me to implement a halting condition which essentially stops the GA from evolving further if all of the keyboards in the new generation have fitness values within one sigma of each other.  When this happens the Evolve function will just return the same generation that was passed into it so no new generations will be produced.

Below is an example of how to use the new glazed.Evolve() function.

```python
newkList = [ (kb, MCFitness(kb)) for kb in klist ]
print("The Standard Qwerty fitness is: ", MCFitness(k).Fitness(), "\n")

averageList = []
bestKeyboard = []
worstKeyboard = []
averageList.append(getAve(newkList))
bestKeyboard.append(getMax(newkList))
worstKeyboard.append(getMin(newkList))

print ("Generation 0")
print ("Average: ", averageList[0])
print ("Maximum Fitness = ", bestKeyboard[0][1].Fitness(), "\nError = ", bestKeyboard[0][1].Error(),"\n")

for i in range(nGen):
    newkListTmp = glazed.Evolve(newkList, fitness=MCFitness, pressurePoint=int(len(newkList)*0.1),mutationRate=0.3)

    #check if the halting condition was met
    if(newkListTmp == newkList):
        print("THE GA HAS BEEN HALTED!!")
        print("There is no discernible difference between the fitness of all the keyboards in the generation")
        break
    else:
        newkList = newkListTmp
    
    #######################################################################################
    # The rest of this is just for plotting useful quantities as the generations progress #
    #######################################################################################

    averageList.append(getAve(newkList))
    bestKeyboard.append(getMax(newkList))
    worstKeyboard.append(getMin(newkList))
    
    print ("Generation ", i+1)
    print ("Average: ", averageList[i+1])
    print ("Maximum Fitness = ", bestKeyboard[i+1][1].Fitness(), "\nError = ", bestKeyboard[i+1][1].Error(),"\n")
    
    bestFitness = [ kb[1].Fitness() for kb in bestKeyboard ]
    bestErrorList = [ kb[1].Error() for kb in bestKeyboard ]
    worstFitness = [ kb[1].Fitness() for kb in worstKeyboard ]
    worstErrorList = [ kb[1].Error() for kb in worstKeyboard ]
    errorListAve = [ kb[1].Error()/sqrt(nGen) for kb in bestKeyboard ]
    genList = [ i for i in range(len(bestFitness)) ]
    
    draw_progress(genList,averageList,errorListAve,bestFitness,bestErrorList,worstFitness,worstErrorList)
    
    citFitness = [kb[1].Fitness() for kb in newkList]
    citFitError = [kb[1].Error() for kb in newkList]
    
    plt.figure()
    plt.errorbar(range(Nk), citFitness, yerr=citFitError, marker='o')
    plt.ylim((0.6,1))
    plt.plot([0, Nk], [averageList[0], averageList[0]], linestyle='--', linewidth=2)
    plt.plot([0, Nk], [averageList[i], averageList[i]], linestyle='--', linewidth=2, color = 'black')
    plt.show()
    
```
