```python
import seaborn as sns
sns.set(style='ticks',palette='Set2')

allEff, spEff, hermEff, monEff, cubEff, modEff = [],[],[],[],[],[]
allErr, spErr, hermErr, monErr, cubErr, modErr = [],[],[],[],[],[]
nFiles = 0
nData = 0
for dat in datList:
    nFiles += 1
    meta, eff = dat.PythonData()
    i = 0
    fast = []
    for keyboard, d in eff:
        nData += 1
        eff = d['spatial'].Fitness()
        err = d['spatial'].Error()
        spEff.append(eff)
        spErr.append(err)
        allEff.append(eff)
        allErr.append(err)
        
        eff = d['hermite'].Fitness()
        err = d['hermite'].Error()
        hermEff.append(eff)
        hermErr.append(err)
        allEff.append(eff)
        allErr.append(err)
        
        eff = d['monotonic'].Fitness()
        err = d['monotonic'].Error()
        monEff.append(eff)
        monErr.append(err)
        allEff.append(eff)
        allErr.append(err)
        
        eff = d['cubic'].Fitness()
        err = d['cubic'].Error()
        cubEff.append(eff)
        cubErr.append(err)
        allEff.append(eff)
        allErr.append(err)
        
        eff = d['modCubic'].Fitness()
        err = d['modCubic'].Error()
        modEff.append(eff)
        modErr.append(err)
        allEff.append(eff)
        allErr.append(err)
        
print(nFiles, "files loaded.")
print(nData, "data points loaded per interpolation.")
```

```python
plt.figure(figsize=(10,8))
#bins = np.linspace(0,100,101)
n1, b1, p1 = plt.hist(allEff,50,normed=False, alpha=0.65)
plt.xlabel("Efficiency (Radix, 5k iterations)",fontsize=15)
plt.show()

plt.figure(figsize=(10,8))
plt.hist(spEff,b1,normed=False,alpha=0.6,label='spatial')
plt.hist(hermEff,b1,normed=False,alpha=0.6,label='hermite')
plt.hist(monEff,b1,normed=False,alpha=0.6,label='monotonic')
plt.hist(cubEff,b1,normed=False,alpha=0.6,label='cubic')
plt.hist(modEff,b1,normed=False,alpha=0.6,label='mod')
plt.legend(loc='upper center')
plt.xlabel("Efficiency (Radix, 5k iterations)",fontsize=15)
plt.show()

plt.figure(figsize=(10,8))
n2, p2, b2 = plt.hist([spEff,hermEff,monEff,cubEff,modEff],50,
                      label=["spatial","hermite","monotonic","cubic","mod cubic"], 
                      normed=False,stacked=True,alpha=0.75)
plt.legend(loc='upper center')
plt.xlabel("Efficiency (Radix, 5k iterations)",fontsize=15)
plt.show()
#savefig('optimizatinDistribution.pdf')
```

```python
spArray = np.asarray(spEff)
hermArray = np.asarray(hermEff)
monArray = np.asarray(monEff)
cubArray = np.asarray(cubEff)
modArray = np.asarray(modEff)

effArrays = []
effArrays.append(np.asarray(spEff))
effArrays.append(np.asarray(hermArray))
effArrays.append(np.asarray(monEff))
effArrays.append(np.asarray(cubEff))
effArrays.append(np.asarray(modEff))

effArray = np.array([spArray,hermArray,monArray,cubArray,modArray], np.float64)
```

```python
nInterps = 5
fontSize = 19
nplt = 1
nyLabel = 0
intNames = ["spatial","hermite","monotonic","cubic","modCubic"]
plt.figure(figsize=(20,20))
for i in range(0,nInterps):
    for j in range(0,nInterps):
        if j == i+1:
            subplot(nInterps,nInterps,nplt)
            xticks([]), yticks([])
            text(0.5,0.01,intNames[nyLabel],ha='center',va='bottom',size=fontSize,alpha=1.0)
            axis('off')
            nplt += 1
            continue
        elif j > i+1:
            nplt += 1
            continue
        ax = plt.subplot(nInterps,nInterps,nplt)
        #ax.set_xlim(0.75,0.95)
        #ax.set_ylim(0.75,0.95)
        sns.kdeplot(effArray[i],effArray[j],shade=False,cmap="Blues")
        sns.regplot(effArray[i],effArray[j],color="SeaGreen",marker="x")
        
        if(nplt <= nInterps):
            ax.set_title(intNames[nplt-1],fontsize=fontSize)
        if(nplt%nInterps == 1):
            ax.set_ylabel(intNames[nyLabel],fontsize=fontSize)
            nyLabel += 1
        plt.tight_layout()
        plt.grid(True)
        nplt += 1
plt.show()
```
