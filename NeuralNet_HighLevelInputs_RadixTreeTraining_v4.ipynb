```python
from donut import *
import pickle
import math
import matplotlib.pyplot as plt
import matplotlib as mpl
from pylab import *
from matplotlib.colors import LogNorm
%matplotlib inline

w = sprinkles.ReadBook("../HuckFinn.txt")
w = sprinkles.MostCommon(w,500)

k = bearclaw.MakeStandardKeyboard()
k.RemoveKey('.')
kStan = k

ivL = 50
Xscale = 0.75
Yscale = 0.5
DDscale = 0.005
```

```python
#######################################################################
##  Utility plotting functions to track NN progress during training  ##
#######################################################################

def drawerrors(e):
    plt.figure(figsize=(4,2))
    x = list(range(len(e)))
    plt.plot(x, e, 'or')
    plt.xlabel('iteration')
    plt.ylabel('error')
    plt.yscale('log')
    plt.show()
    
    
def draw(testSet,k,randomize=False):
    x,y,yerr,acc = [],[],[],[]
    acc = 0
    for j in range(len(testSet)):
        testV = []
        word1 = str(testSet[j][0])
        word2 = str(testSet[j][1])
        if randomize == True:
            k = bearclaw.RandomKeyboard(k)    
        iv = NNmodel.RandomVector(word1,k)
        
        testV = NNmodel.GetInputVector(iv,word2,k)                
        predict = net.activate(testV)
        x.append(predict[0])
        y.append(testSet[j][2])
        yerr.append(0)

        if predict > 0.5 and testSet[j][2] == 1:
            acc += 1
        elif predict <= 0.5 and testSet[j][2] == 0:
            acc += 1     
        
    plt.errorbar(x,y,yerr=yerr,fmt='ob',color='green')    
    plt.ylim(-0.1,1.1)
    plt.title("Correlation")
    plt.xlabel('NN Prediction')
    plt.ylabel('Truth')
    plt.show()
    print("Accuracy = ", acc/len(testSet))
    
def drawTest(testSet,k,randomize=False):
    x,y,yerr,acc = [],[],[],[]
    acc = 0
    for j in range(len(testSet)):
        testV = []
        word1 = str(testSet[j][0])
        word2 = str(testSet[j][1])
        if randomize == True:
            k = bearclaw.RandomKeyboard(k)    
        iv = NNmodel.RandomVector(word1,k)
        
        testV = NNmodel.GetInputVector(iv,word2,k)                
        predict = net.activate(testV)
        x.append(predict[0])
        y.append(testSet[j][2])
        yerr.append(0)

        if predict > 0.5 and testSet[j][2] == 1:
            acc += 1
        elif predict <= 0.5 and testSet[j][2] == 0:
            acc += 1     
        
    print("Accuracy = ", acc/len(testSet))
    plt.errorbar(x,y,yerr=yerr,fmt='ob',color='green')    
    plt.ylim(-0.1,1.1)
    plt.title("Correlation")
    plt.xlabel('NN Prediction')
    plt.ylabel('Truth')
    plt.show()
    
    xedges = []
    nXbins = 100
    for i in range(nXbins):
        xedges.append(i/nXbins)
    xedges.append(1)
    yedges = [-0.01,0.5,1.01]
    hist2d(x, y, norm=LogNorm(),bins=(xedges, yedges))
    colorbar()
```

```python
#################################
##  Neural Net Training Class  ##
#################################

class NN_HLIv3(cruller.InputModel):
    def __init__(self, scale = 1, vectorlength = ivL, corr = 0.0):
        cruller.InputModel.__init__(self) #don't forget this line
        
        self.model = cruller.SimpleInterpolationModel()
        self.model.SetVectorLength(vectorlength)
        self.model.SetXScale(Xscale)
        self.model.SetYScale(Yscale)
        self.model.SetCorrelation(corr)
        self.perfectmodel = cruller.SimpleInterpolationModel()
        self.perfectmodel.SetScale(0.0)
        self.perfectmodel.SetVectorLength(vectorlength)
        
        
    def RandomVector(self, word, keyboard):
        #you can add noise to this or use other interpolation methods here          
        return cruller.SpatialInterpolation(self.model.RandomVector(word,keyboard), ivL)
    
    
    def PerfectVector(self, word, keyboard):
        #you can add noise to this or use other interpolation methods here        
        return cruller.SpatialInterpolation(self.perfectmodel.RandomVector(word,keyboard), ivL)
    
    
    def GetInputVector(self,iv,word,keyboard):
        ivPoints = iv.PointList()  
        iv2 = self.perfectmodel.RandomVector(word,keyboard)                      
        ivPoints2 = iv2.PointList()    
        Xdist, DXdist, D2Xdist = 0,0,0
        Ydist, DYdist, D2Ydist = 0,0,0        
        #compute the derivatives
        DivPoints,DivPoints2 = [],[]          
        for i in range(0,min(len(ivPoints),len(ivPoints2))):
            if i == 0:
                DivPoints.append((0,0))
                DivPoints2.append((0,0))                            
                Xdist += (ivPoints[i][0] - ivPoints2[i][0])**2
                Ydist += (ivPoints[i][1] - ivPoints2[i][1])**2
            else:
                Xdist += (ivPoints[i][0] - ivPoints2[i][0])**2
                Ydist += (ivPoints[i][1] - ivPoints2[i][1])**2            
                Dx1 = ivPoints[i][0]-ivPoints[i-1][0]
                Dy1 = ivPoints[i][1]-ivPoints[i-1][1]
                Dx2 = ivPoints2[i][0]-ivPoints2[i-1][0]
                Dy2 = ivPoints2[i][1]-ivPoints2[i-1][1]                                            
                DivPoints.append((Dx1,Dy1))
                DivPoints2.append((Dx2,Dy2))            
                DXdist += (Dx1 - Dx2)**2
                DYdist += (Dy1 - Dy2)**2            
                D2x1 = DivPoints[i][0]-DivPoints[i-1][0]
                D2y1 = DivPoints[i][1]-DivPoints[i-1][1]
                D2x2 = DivPoints2[i][0]-DivPoints2[i-1][0]
                D2y2 = DivPoints2[i][1]-DivPoints2[i-1][1]
                D2Xdist += (D2x1 - D2x2)**2
                D2Ydist += (D2y1 - D2y2)**2
        #Create Input Vector for NN
        
        #normalized spatial difference of swype vectors
        inputVect = []
        l1 = iv.SpatialLength()
        l2 = iv2.SpatialLength()        
        inputVect.append(abs(l1-l2)/(max(l1,l2) if max(l1,l2) else 1))                                                                    
        #x and y distances
        inputVect.append(Xdist)
        inputVect.append(Ydist)
        #x and y first derivative distances
        inputVect.append(DXdist)
        inputVect.append(DYdist)
        #x and y second derivative distances
        inputVect.append(D2Xdist)
        inputVect.append(D2Ydist)
        #x and y distance between initial points
        inputVect.append((iv.PointList()[0][0]-iv2.PointList()[0][0])**2)
        inputVect.append((iv.PointList()[0][1]-iv2.PointList()[0][1])**2)
        #x and y distance between the last points
        inputVect.append((iv.PointList()[ivL-1][0]-iv2.PointList()[ivL-1][0])**2)
        inputVect.append((iv.PointList()[ivL-1][1]-iv2.PointList()[ivL-1][1])**2)                
        return inputVect
    
    
NNmodel = NN_HLIv3()
```

```python
##############################
##  Set Up Neural Net (NN)  ##
##############################

from pybrain.structure import FeedForwardNetwork, LinearLayer, SigmoidLayer, FullConnection, BiasUnit
from pybrain.structure.modules.neuronlayer import NeuronLayer
from pybrain.supervised.trainers import RPropMinusTrainer

nInputs = 11
net = FeedForwardNetwork()

#Define Layers
inLayer = LinearLayer(nInputs)
net.addInputModule(inLayer)

hLayer1 = SigmoidLayer(nInputs)
net.addModule(hLayer1)

hLayer2 = SigmoidLayer(nInputs)
net.addModule(hLayer2)

hLayer3 = SigmoidLayer(nInputs)
net.addModule(hLayer3)

hLayer4 = SigmoidLayer(2)
net.addModule(hLayer4)

outLayer = SigmoidLayer(1)
net.addOutputModule(outLayer)

hBiasLayer = BiasUnit()
net.addModule(hBiasLayer)

#Define Connections
in_to_h1 = FullConnection(inLayer,hLayer1)
net.addConnection(in_to_h1)

h1_to_h2 = FullConnection(hLayer1,hLayer2)
net.addConnection(h1_to_h2)

h2_to_h3 = FullConnection(hLayer2,hLayer3)
net.addConnection(h2_to_h3)

h3_to_h4 = FullConnection(hLayer3,hLayer4)
net.addConnection(h3_to_h4)

h4_to_out = FullConnection(hLayer4,outLayer)
net.addConnection(h4_to_out)

net.addConnection(FullConnection(hBiasLayer,hLayer1))
net.addConnection(FullConnection(hBiasLayer,hLayer2))
net.addConnection(FullConnection(hBiasLayer,hLayer3))
net.addConnection(FullConnection(hBiasLayer,hLayer4))
net.addConnection(FullConnection(hBiasLayer,outLayer))

net.sortModules()
```

```python
##############################
##  Set Up NN Pre-Training  ##
##############################

f = open('WordPairTrainingSet.dat','rb')
preTrainingSet = pickle.load(f)

from pybrain.datasets import SupervisedDataSet
preDS = SupervisedDataSet(nInputs, 1)

preTrainSize = 50000
for i in range(len(preTrainingSet[:preTrainSize])):
    inputVect = []
    word1 = str(preTrainingSet[i][0])
    word2 = str(preTrainingSet[i][1])
    k = bearclaw.RandomKeyboard(k)    
    iv = NNmodel.RandomVector(word1,k)
    
    inputVect = NNmodel.GetInputVector(iv,word2,k)
    result = preTrainingSet[i][2]
    preDS.addSample(inputVect,result)
```

```python
#######################
##  NN Pre-training  ##
#######################

preTrainIt = 25
preTrainer = RPropMinusTrainer(net)
preTrainer.setData(preDS)

preTestSize = 100
preTestSet = preTrainingSet[len(preTrainingSet)-preTestSize:]

errs = []
for i in range(preTrainIt):
    e = preTrainer.train()
    draw(preTestSet,k,randomize=True)
    print(e)
    errs.append(e)
    drawerrors(errs)
```
