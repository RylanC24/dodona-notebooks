```python
from donut import *
import pickle
from math import *
import matplotlib.pyplot as plt
import matplotlib as mpl
from pylab import *
from matplotlib.backends.backend_pdf import PdfPages 
%matplotlib inline

import pandas as pd
import seaborn as sns
cp = ["#c51b7d","#f1b6da","#4d9221"]
sns.palplot(sns.color_palette(cp, 3))
sns.set(style='ticks',palette=cp)
colorm="#c51b7d"
colorg="#4d9221"

w = cruller.WordList()
w.LoadFromFile("/Users/Rylan/cisc/creamTeam/wordLists/WordList_100000_combined_check.dat")
```

```python
cumTotal = 0
cumDist = []
wOccurances = []
cumTotal20k = 0
for i in range(w.Words()):
    wOccurances.append(w.Occurances(i))
    cumTotal += w.Occurances(i)
    cumDist.append(cumTotal)
    if i < 20000:
        cumTotal20k += w.Occurances(i)
        
wFrequencies = [float(wOccurances[i]/cumTotal) for i in range(len(wOccurances))]
cumDist = [float(cumDist[i]/cumTotal) for i in range(len(cumDist))]
bins = linspace(0,w.Words(),w.Words()+1)
print(cumTotal, cumTotal20k, cumTotal20k/cumTotal)
```

```python
lfs = 19
tfs = 15

fig, ax1 = plt.subplots(figsize=(10,7))
ax1.axvline(x=20000,color='black',linestyle="--")
ax1.plot(wFrequencies,color=colorm,marker='.',ls='',rasterized=True)
ax1.set_xlabel("$N^{th}$ Most Common Word",fontsize=lfs)
ax1.set_ylabel("Individual Word Frequency",fontsize=lfs,color=colorm)
ax1.set_xscale("log")
ax1.set_yscale("log")
ax1.xaxis.labelpad = 13
for tl in ax1.get_xticklabels():
    tl.set_size(tfs)
for tl in ax1.get_yticklabels():
    tl.set_color(colorm)
    tl.set_size(tfs)
    
ax2 = ax1.twinx()
ax2.plot(cumDist,color=colorg,lw=1,ls='',marker='.',rasterized=True)
ax2.set_ylabel("Cumulative Word Frequency",fontsize=lfs,color=colorg)
ax2.yaxis.labelpad = 15
for tl in ax2.get_yticklabels():
    tl.set_color(colorg)  
    tl.set_size(tfs)

ax3=ax1.twiny()
ax3.plot(wFrequencies,color=colorm,marker='.',ls='',rasterized=True)
ax3.set_xscale("log")
ax3.set_xticklabels([])
savefig('/Users/Rylan/cisc/creamTeam/DonutDocs/plots/finalPlots/wordList_freqDist.svg')
plt.show()
```
