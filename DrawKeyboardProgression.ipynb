```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pylab import savefig
%matplotlib inline
import random

#Donut stuff
from donut import *
random.seed()
randSeed = random.randint(0,1000000)

w = cruller.WordList()
w.LoadFromFile("/Users/Rylan/cisc/creamTeam/wordLists/WordList_20000_combined_check.dat")
w.SetSeed(randSeed)
```

```python
#Load the connections from the nn 
file= open("/Users/Rylan/cisc/creamTeam/donutData/finalOptData/gaussOutput/onlyImprovedKeyboards.txt","r")
data = [str(line.split()[0]) for line in file]

kbdStrings = []
prevString = ""
for d in data:
    if d != prevString:
        kbdStrings.append(d)
    prevString = d

keyboards = [bearclaw.MakeStandardKeyboard(alphabetStr=kstring+'.') for kstring in kbdStrings]
[kbd.RemoveKey('.') for kbd in keyboards]
nKeyboards = len(keyboards)
print("nKeyboards = ", nKeyboards)
```
